```python
new_meta = meta.loc[meta.year == 2017,['genres','id','title','year']]
new_meta
```

```python
new_meta['id'] = new_meta['id'].astype(int)
```

```python
data = pd.merge(new_meta, credits, on='id')
data
```

```python
import ast
data['genres'] = data['genres'].map(lambda x: ast.literal_eval(x))
data['cast'] = data['cast'].map(lambda x: ast.literal_eval(x))
data['crew'] = data['crew'].map(lambda x: ast.literal_eval(x))
```

```python
def make_genresList(x):
    gen = []
    st = " "
    for i in x:
        if i.get('name') == 'Science Fiction':
            scifi = 'Sci-Fi'
            gen.append(scifi)
        else:
            gen.append(i.get('name'))
    if gen == []:
        return np.NaN
    else:
        return (st.join(gen))
```

```python
data['genres_list'] = data['genres'].map(lambda x: make_genresList(x))
data['genres_list'][:15]
```

```python
data['cast'][0]
```

```python
def get_act(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == []:
        return np.NaN
    else:
        return casts[0]
```

```python
def get_act2(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=1:
        return np.NaN
    else:
        return (casts[1])
```

```python
def get_act3(x):
    casts = []
    for i in x:
        casts.append(i.get('name'))
    if casts == [] or len(casts)<=2:
        return np.NaN
    else:
        return (casts[2])
```

```python
data['actor_1_name']= data['cast'].map(lambda x: get_act(x))
data['actor_2_name'] = data['cast'].map(lambda x: get_act2(x))
data['actor_3_name'] = data['cast'].map(lambda x: get_act3(x))
print(data['actor_1_name'][:15])
print(data['actor_2_name'][:15])
print(data['actor_3_name'][:15])
```

```python
def get_direct(x):
    dt = []
    st = " "
    for i in x:
        if i.get('job') == 'Director':
            dt.append(i.get('name'))
    if dt == []:
        return np.NaN
    else:
        return (st.join(dt))
```

```python
data['director_name'] = data['crew'].map(lambda x: get_direct(x))
data['director_name'][:15]
```
